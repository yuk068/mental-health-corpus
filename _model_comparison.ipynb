```python
import matplotlib.pyplot as plt
import numpy as np

# Model names and F1 scores
models = ['M NB (BOW)', 'G NB (LSA)', 'Euclidean K-NN K=50 (TF-IDF)', 'Cosine K-NN K=10 (LSA)', 
          'RF (TF-IDF)', 'RF (LSA)', 'L2 Log.Reg. C=5 (TF-IDF)', 'L2 Log.Reg. C=5 (LSA)', 
          'SoftM SVM C=0.1 (TF-IDF)', 'SoftM SVM C=10 (LSA)', 'RBF SVM C=100 G=0.1 (LSA)']
f1_scores = [0.875, 0.777, 0.887, 0.865, 0.890, 0.889, 0.932, 0.918, 0.924, 0.910, 0.916]

# Reverse the order of the models and F1 scores
models.reverse()
f1_scores.reverse()

# Create the plot
plt.figure(figsize=(10, 6))
bars = plt.barh(models, f1_scores, color='skyblue')

# Highlight the best model (maximum F1 score)
best_model_idx = f1_scores.index(max(f1_scores))
bars[best_model_idx].set_color('orange')

# Add F1 score labels inside the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.02, bar.get_y() + bar.get_height() / 2, f'{width:.3f}', va='center', ha='right', color='black')

# Set labels and title
plt.xlim(0.7, 1.0)
plt.xlabel('F1 Score')
plt.title('Model F1 Scores Comparison')

# Add horizontal gridlines
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Subdivide xlim further
x_ticks = np.arange(0.7, 1.0, 0.03)  # Create ticks from 0.7 to 1.0 with step size of 0.05
plt.xticks(x_ticks)

plt.tight_layout()

# Show the plot
plt.show()
```
