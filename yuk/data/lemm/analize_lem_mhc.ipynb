```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.tokenize import word_tokenize
import nltk

# Ensure you have the required resources
# nltk.download('punkt')

# Load your dataset
df = pd.read_csv('lem_mhc_updated_2.csv')

# Function to tokenize text and count word frequencies
def get_word_frequencies(text_series):
    all_words = []
    for text in text_series:
        words = word_tokenize(text.lower())  # Tokenize and convert to lowercase
        all_words.extend(words)
    return Counter(all_words)

# Separate the text by label
text_label_1 = df[df['label'] == 1]['text']
text_label_0 = df[df['label'] == 0]['text']

# Get word frequencies for each label
word_freq_label_1 = get_word_frequencies(text_label_1)
word_freq_label_0 = get_word_frequencies(text_label_0)

# Get the top 30 words for each label
common_words_label_1 = word_freq_label_1.most_common(30)
common_words_label_0 = word_freq_label_0.most_common(30)

# Convert to DataFrame for easier plotting
df_label_1 = pd.DataFrame(common_words_label_1, columns=['Word', 'Frequency'])
df_label_0 = pd.DataFrame(common_words_label_0, columns=['Word', 'Frequency'])

# Plot the top 30 most common words for label 1
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Word', data=df_label_1, palette='Blues_d')
plt.title('Top 30 Most Common Words in Label 1')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()

# Plot the top 30 most common words for label 0
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Word', data=df_label_0, palette='Greens_d')
plt.title('Top 30 Most Common Words in Label 0')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from nltk import ngrams
import nltk

# Ensure you have the required resources
# nltk.download('punkt')

# Load your dataset
df = pd.read_csv('lem_mhc_updated_2.csv')

# Function to generate n-grams and count their frequencies
def get_ngram_frequencies(text_series, n):
    all_ngrams = []
    for text in text_series:
        words = word_tokenize(text.lower())  # Tokenize and convert to lowercase
        ngrams_list = ngrams(words, n)  # Generate n-grams
        all_ngrams.extend(ngrams_list)
    return Counter(all_ngrams)

# Separate the text by label
text_label_1 = df[df['label'] == 1]['text']
text_label_0 = df[df['label'] == 0]['text']

# Get bigram and trigram frequencies for each label
bigram_freq_label_1 = get_ngram_frequencies(text_label_1, 2)
bigram_freq_label_0 = get_ngram_frequencies(text_label_0, 2)
trigram_freq_label_1 = get_ngram_frequencies(text_label_1, 3)
trigram_freq_label_0 = get_ngram_frequencies(text_label_0, 3)

# Get the top 15 bigrams and trigrams for each label
top_bigrams_label_1 = bigram_freq_label_1.most_common(15)
top_bigrams_label_0 = bigram_freq_label_0.most_common(15)
top_trigrams_label_1 = trigram_freq_label_1.most_common(15)
top_trigrams_label_0 = trigram_freq_label_0.most_common(15)

# Convert to DataFrame for easier plotting, converting tuples to strings
df_bigrams_label_1 = pd.DataFrame([(str(bigram), freq) for bigram, freq in top_bigrams_label_1], columns=['Bigram', 'Frequency'])
df_bigrams_label_0 = pd.DataFrame([(str(bigram), freq) for bigram, freq in top_bigrams_label_0], columns=['Bigram', 'Frequency'])
df_trigrams_label_1 = pd.DataFrame([(str(trigram), freq) for trigram, freq in top_trigrams_label_1], columns=['Trigram', 'Frequency'])
df_trigrams_label_0 = pd.DataFrame([(str(trigram), freq) for trigram, freq in top_trigrams_label_0], columns=['Trigram', 'Frequency'])

# Plotting the top 15 bigrams for label 1
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Bigram', data=df_bigrams_label_1, palette='Blues_d')
plt.title('Top 15 Bigrams in Label 1')
plt.xlabel('Frequency')
plt.ylabel('Bigrams')
plt.show()

# Plotting the top 15 bigrams for label 0
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Bigram', data=df_bigrams_label_0, palette='Greens_d')
plt.title('Top 15 Bigrams in Label 0')
plt.xlabel('Frequency')
plt.ylabel('Bigrams')
plt.show()

# Plotting the top 15 trigrams for label 1
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Trigram', data=df_trigrams_label_1, palette='Blues_d')
plt.title('Top 15 Trigrams in Label 1')
plt.xlabel('Frequency')
plt.ylabel('Trigrams')
plt.show()

# Plotting the top 15 trigrams for label 0
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Trigram', data=df_trigrams_label_0, palette='Greens_d')
plt.title('Top 15 Trigrams in Label 0')
plt.xlabel('Frequency')
plt.ylabel('Trigrams')
plt.show()
```
